# Step 4: Target Variable Create Karna
Hum apne target variable Pass/Fail ko banayenge. Hum Pass ko 1 aur Fail ko 0 assign karenge, based on the total score of the student. Agar student ka total score 150 ya zyada hai to wo Pass hoga.

- Total Score Calculate Karna: Hum math, reading aur writing scores ko add karenge.

- Pass/Fail Column Banana: Agar total score 150 ya zyada ho to Pass (1), warna Fail (0).

Explanation:

- df['Total Score'] = df['math score'] + df['reading score'] + df['writing score']: Is line mein hum teen scores ko add kar rahe hain, taake student ka total score mile.

- apply(lambda x: 1 if x >= 150 else 0): Yeh logic hai jo total score ke basis par Pass/Fail decide karta hai. Agar total score 150 ya zyada hai to wo 1 (Pass) hoga, agar kam hai to 0 (Fail) hoga.

```python
# Total score calculate kar rahe hain (math + reading + writing)
df['Total Score'] = df['math score'] +df['reading score']+df['writing score']

# Pass/Fail target variable create kar rahe hain
df['Pass/Fail'] = df['Total Score'].apply(lambda x:1 if x>=150 else 0 ) 
```

# Step 5 : Categorical Features Ko Encode Karna

- Ab hum categorical variables ko numerical format mein convert karenge. Jaise ke gender, race/ethnicity, lunch, aur test preparation ko numbers mein badalna.

- Hum One-Hot Encoding ka use karenge jisme har category ko ek alag column milta hai. Jaise agar gender hai to male aur female ke liye alag columns banenge.

Explanation:

- pd.get_dummies(df, columns=[...], drop_first=True): Yeh line categorical variables ko numerical format mein convert kar rahi hai. One-Hot Encoding use ho raha hai.

- drop_first=True: Isse pehli category ko drop kar diya jata hai, taake hum multicollinearity se bach sakein.

Ab gender, race, lunch, aur test preparation ko numerical values mil gayi hain, aur ab hum in features ko model mein use kar sakte hain.

```python
# Categorical columns ko numerical format mein convert kar rahe hain
df = pd.get_dummies(df, columns=['gender', 'race/ethnicity', 'lunch', 'test preparation course'], drop_first=True)

# Encoded data ka preview dekhte hain
df.head()
```
